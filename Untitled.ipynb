```python
import requests
import pprint
import os
from bs4 import BeautifulSoup
from woocommerce import API
import pandas as pd

session = requests.Session()

print('GardenScraper v.2 by Django Claughan \nEdited by Fatih Canli' )

# Get login details form setup file
#thisFolder = os.path.dirname(os.path.abspath(__file__))
#setupFile = os.path.join(thisFolder, 'SETUP.txt')
f = open('SETUP.txt', "r")
details = f.read().split('\n') 

links = open('links.txt', "r")
link = links.read().split('\n') 

link = [x for x in link if x.find('https://www.gardners.com/Product/9') is 0]

print('Gardeners details')
print(details[4])
print(details[5])
print(details[6])
print('')
print('WooCommerce details')
print(details[9])
print(details[12])
print(details[15])
print(str(len(link))+' entries detected. Please press ENTER to proceed')
sheet='Full Entries'
excel_file='output.xlsx'
def excel_reader(excelf,sheet):
    df0=pd.read_excel(excelf,sheet_name=sheet)
    data=[]
    for i in range (len(df0)):  
    
        new_data = {
        'images': [{
            "src": df0['image'][i]
        },
                ],
        'name': df0['name'][i],
        'regular_price': df0['regular_price'][i],
        'short_description': df0['description'][i],
        'sku': str(df0['isbn'][i]),
        'manage_stock': True,
        'stock_quantity': int(df0['stock'][i]),
        'status': 'publish',
        'type': 'simple'
        }
        data.append(new_data)
    return data


input()


# Initialise Woocommerce
wcapi = API(
    url = details[9],
    consumer_key = details[12],
    consumer_secret = details[15],
    version = 'wc/v3'
)

# Package Gardeners login details
payload = {
    'AccountNumber': details[4], 
    'UserName': details[5],
    'Password': details[6]
}

# Post the payload to the site to log in

log=[]
s = session.post("https://www.gardners.com/Account/LogOn", data=payload)

print()

for url in link:
# Start loop to add products

  # Navigate to the next page and scrape the data
  page = session.get(url)

  soup = BeautifulSoup(page.content, 'html.parser')
  results = soup.find(id='body')
  # Get title
    
  searchBlock = results.find('div', class_='titleContributor')

  if searchBlock == None:
        print(url)
        print('No data retrieved.')
        print('\n\n')
        title='NO DATA'
        rrp='NO DATA'
        isbn='NO DATA'
        stock = 'NO DATA'
        img='NO DATA'
        description1='NO DATA'
  else: 
          title = searchBlock.find('h1')
          title = (title.text) 
            
          # Get ISBN
          searchBlock = results.find('li', class_='isbn')
          if searchBlock == None:
                isbn ='No ISBN'
          else:
                isbn = searchBlock.find_all('span')
                isbn = isbn[1].text
        
          # Get RRP
          searchBlock1 = results.find('div', class_='purchaseBlock')
          rrp = searchBlock1.find('span', class_='retailPrice')
          if rrp == None:
                print(url)
                rrp = searchBlock1.find('p', class_='rrp')
                if rrp == None:
                   print(isbn+' '+title+' '+' has no RRP ')
                   rrp1='NOT OK'
                else:
                   print(isbn+' '+title+' '+'is available.')
                   rrp = rrp.text
                   rrp = rrp.replace('RRP:', '')
                   rrp1='OK'
                   print(rrp)
                
          else: 
                  print(isbn+' '+title+' '+'is available.')
                  rrp = rrp.text
                  rrp1='OK'
                  print(rrp)
          # Get author(s)
          searchBlock = results.find('div', class_='titleContributor')
          title = searchBlock.find('h1')
          title = (title.text) 
          author = searchBlock.find_all('a')
          # Add author(s) to title
          n = 0
          for i in author:
              if n == 0:
                  title = title + ' by ' + i.text
              elif n == len(author) - 1:
                  title = title + ' and ' + i.text
              else:
                  title = title + ', ' + i.text
              n = n + 1
          print(title)  

                
          # Get stock and change number for site
          searchBlock1 = results.find('div', class_='purchaseBlock')
          
          try:
              stock = searchBlock1.find(class_='availability')['data-copies']
          except:
              stock = 1
        
          if int(stock) > 4:
              stock = 1
          elif int(stock) > 0:
              stock = stock
          else:
              stock = 1
          print('Stock: '+str(stock))
          

          # Get description
          searchBlock = results.find('div', class_='description')
          if searchBlock == None:
                description = ''
                description1 = 'NOT OK'
                print('No description')
          else:
                description = searchBlock.find(class_='productDescription')
                description = description.text
                description1 = 'OK'
          # Get image url
          searchBlock = results.find(class_='imageContainer')

          if searchBlock.find('img')['src'] == "/Public/Images/noimage.jpg":
                imageSrc = 'https://www.narberthmuseum.co.uk/wp-content/uploads/woocommerce-placeholder.png'
                print('No image available')
                img='NOT OK'
          else:
                imageSrc = searchBlock.find('img')['data-zoom-image']
                imageSrc = 'https:' + imageSrc
                img='OK'

          

  log_data = {
      'url' : url,
      'isbn':isbn,
      'name':title,
      'regular_price':rrp,
      'image':imageSrc,
      'stock':stock,
      'description':description,
      'has_img':img,
      'has_desc':description1,
       'has_rrp':rrp1
       }
  log.append(log_data)

    
df = pd.DataFrame.from_dict(log)
mask=(df['has_desc']=='OK') & ( df['has_img']=='OK')  & ( df['has_rrp']=='OK')
mask1=(df['has_desc']=='NOT OK') | ( df['has_img']=='NOT OK') | ( df['has_rrp']=='NOT OK')
df2=df[mask]
df1=df[mask1]
df2.index+=1
df1.index+=1

writer = pd.ExcelWriter('output.xlsx', engine='xlsxwriter')
df2.to_excel(writer,sheet_name='Full Entries')
df1.to_excel(writer,sheet_name='Missing Entries')
writer.save()
    
# Package data to post to Woocommerce
data=excel_reader(excel_file,sheet)

print('Process completed. Press ENTER to proceed to post data')

input()        
          # Post data to the Woocommerce API

for each_data in data:
    print(wcapi.post('products',each_data).json())
    print('')



print('Program finished. Press ENTER to close')

input()
```
